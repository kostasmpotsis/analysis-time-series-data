```python
amol95 = '((α|a)?μ(ό|ο|o).*95.*)'
amol100 = '(Αμό.*100.*)'
sup = '(S?uper.*)'
autogas = '(.*(a|A)utogas.*)'
diesel_kin="(Diesel\s+(Κ|κ).*)"
diesel_ther="(Diesel\s+Θ.*)"
diesel_ther2 = '(Diesel\s+Θ(έ|ε)ρμανσης\n\d[.]\d+,\d+)'
```

```python
df["product"]=df['product'].str.strip().str.replace(amol95,'Αμόλυβδη 95', regex=True,case=False)
df["product"]=df['product'].str.replace(amol100,'Αμόλυβδη 100', regex=True)
df["product"]=df['product'].str.replace(sup,'Super', regex=True)
df["product"]=df['product'].str.replace(diesel_kin,'Diesel Kίνησης', regex=True, case=False)
df["product"]=df['product'].str.replace(diesel_ther,'Diesel Θέρμανσης', regex=True, case=False)
df["product"]=df['product'].str.replace(diesel_ther2,'Diesel Θέρμανσης', regex=True, case=False)
df["product"]=df['product'].str.replace(autogas,'Υγραέριο κίνησης', regex=True, case=False)
df['product'] = df['product'].str.replace('Προϊόν','nan')
df['product'] = df['product'].str.replace('σημειωσεις','nan',case=False)
df['product'] = [get_nan(product) for product in df['product']]
df["product"].value_counts(dropna=False)
```
