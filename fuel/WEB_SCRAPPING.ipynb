```python
response = requests.get('http://www.fuelprices.gr/deltia_d.view') 
soup = BeautifulSoup(response.text, 'html.parser')
# soup
```

```python
links = [a['href'] for a in soup.select('a[href]')]

df = pd.DataFrame(links)
df = df.iloc[6:,:].reset_index().drop(['index'],axis=1)
df = df.rename(columns={df.columns[0]:'file_href'})

garbage = '(^(./))'
base = 'http://www.fuelprices.gr/'
df['file_href_cleaned'] = df.file_href.str.replace(garbage,'',regex=True).str.strip()
df['urls'] = base+df.file_href_cleaned

df['date'] = df.file_href.str.extract(r'PANELLINIO_(.*).pdf')
# df['date'] = df.date.str.replace('_','/')
# df['date'] = pd.to_datetime(df['date'], format="%d/%m/%Y").dt.strftime('%Y-%m-%d')

df.head(10)
```

```python
s = 0
for i, item in df.iterrows():
    s += 1
    print("Downloading: File ", i, item['date'], item['urls'])
    print('--------------------------')
    
    time.sleep(randint(1,6))
    response = requests.get(item['urls'])
    
    # Write content in pdf file
    pdf = open(str(item['date'])+".pdf", 'wb')
    pdf.write(response.content)
    pdf.close()
    print("File ", i, " downloaded")
```
